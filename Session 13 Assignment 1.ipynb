```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import sqlite3 as db
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
print(pysqldf)
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
```

```python
df = pd.read_csv(url,sep = ',')
```

```python
df.columns=['age','workclass','fnlwgt','education','educationnum','maritalstatus','occupation','relationship','race','sex','capitalgain','capitalloss','hoursperweek','nativecountry','salary']
```

```python
type(df)
```

```python
#1. Select 10 records from the adult sqladb
pysqldf("select * from df LIMIT 10")
```

```python
#2. Show me the average hours per week of all men who are working in private sector
q = """ select AVG(hoursperweek),sex,workclass from df where TRIM(sex) = 'Male' and TRIM(workclass)='Private' group by sex, workclass;  """
sqldf(q)
```

```python
#3. Show me the frequency table for education, occupation and relationship, separately
#q = """ select education, avg(education),min(education),max(education) from df group by education UNION ALL select occupation,
#avg(occupation),min(occupation),max(occupation) from df group by occupation; """
q = """ select * from df group by education ,occupation, relationship; """
sqldf(q).describe().unstack()
```

```python
#4. Are there any people who are married, working in private sector and having a masters degree

q = """ select * from df where TRIM(maritalstatus) like 'Married%' and TRIM(workclass)='Private' and TRIM(education)='Masters';"""
sqldf(q)
```

```python
#5. What is the average, minimum and maximum age group for people working in different sectors
q = """ select avg(age),min(age), max(age), TRIM(workclass) from df group by workclass; """

sqldf(q)
```
